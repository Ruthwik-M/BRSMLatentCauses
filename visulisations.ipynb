```python
import pandas as pd
```

```python
observation = pd.read_excel("Observation.xlsx")
estimation = pd.read_excel("Estimation.xlsx")
```

```python
generous_obs =  observation[observation.condition == "sparse_above_average"]
generous_obs
generous_obs_test = generous_obs[generous_obs.sequence_type == "test"]
generous_obs_train = generous_obs[generous_obs.sequence_type == "training"]
generous_obs_test
generous_obs_train
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

generous_obs_train_sequence = generous_obs_train.head(40)
sb.histplot(generous_obs_train_sequence.observed_amount, bins= 20)
#plt.savefig("obs_gen_train_seq1.jpg")
```
